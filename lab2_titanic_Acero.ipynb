```python
import pandas as pd
df = pd.read_csv("titanic_cleaned.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

sibsp_pclass_analysis = df.groupby("pclass")["sibsp"].mean().reset_index()

plt.figure(figsize=(7, 5))
plt.bar(sibsp_pclass_analysis["pclass"], sibsp_pclass_analysis["sibsp"], color=["red", "blue", "green"])
plt.title("Average Number of Siblings/Spouses per Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average SibSp")
plt.xticks([1, 2, 3])
plt.show()
```

**ANALYSIS 1**

The Titanic's passenger list reveals a significant relationship between family presence and social class. First-class passengers tended to travel in smaller family units, often as wealthy couples or solo individuals, while third-class passengers frequently traveled in larger family groups. This highlights the socio-economic divide, where wealthier individuals had more independence in their travels, while lower-class families migrated together for economic reasons.

```python
import matplotlib.pyplot as plt

embarked_survival_analysis = df.groupby("embarked")["survived"].mean().reset_index()

plt.figure(figsize=(7, 5))
plt.bar(embarked_survival_analysis["embarked"], embarked_survival_analysis["survived"], color=["blue", "green", "orange"])
plt.title("Survival Rate by Embarkation Point")
plt.xlabel("Port of Embarkation")
plt.ylabel("Survival Rate")
plt.xticks(["C", "Q", "S"], ["Cherbourg", "Queenstown", "Southampton"])
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

**ANALYSIS 2**

The port where passengers boarded the Titanic appears to have influenced their chances of survival. Those who embarked from Cherbourg (C) had the highest survival rate. In contrast, passengers from Southampton (S) had the lowest survival rate. Meanwhile, those from Queenstown (Q) had a moderate survival rate, though their numbers were smaller.
